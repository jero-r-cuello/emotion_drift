```python
      
import pandas as pd
import os
import glob

json_directory = "/home/jcuello/emotion_drift/data/01_stimuli/andyzou_situations_dataset"
```

```python
json_files = glob.glob(os.path.join(json_directory, '*.json'))

print("Found the following JSON files:")
for file in json_files:
    print(f"- {file}")
```

```python
all_dataframes = []

for file_path in json_files:
    filename = os.path.basename(file_path)
    emotion = os.path.splitext(filename)[0]
    
    df = pd.read_json(file_path)
    df.columns = ['situation']
    df['emotion'] = emotion
        
    all_dataframes.append(df)

final_df = pd.concat(all_dataframes, ignore_index=True)
    
final_df.shape
```
