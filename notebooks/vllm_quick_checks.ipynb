```python
import os
import vllm

vllm_path = os.path.dirname(vllm.__file__)
print(f"vllm path: {vllm_path}")
```

```python
from transformers import AutoConfig
MODEL_NAME = "microsoft/Phi-3-medium-128k-instruct"

config = AutoConfig.from_pretrained(MODEL_NAME)
num_layers = config.num_hidden_layers

print(f'The model {MODEL_NAME} has {num_layers} layers.')
```

```python
TARGET_LAYERS = [l for l in range(0,num_layers,2)]
TARGET_LAYERS
```

```python
dimensionality = config.hidden_size

print(f'The model {MODEL_NAME} has a hidden size of {dimensionality}.')
```
